```python
import pandas as pd

# Load the Excel file to inspect its structure
file_path = "data/Série Histórica - Tuberculose.xlsx"
excel_data = pd.ExcelFile(file_path)

# Display sheet names to understand the file structure
excel_data.sheet_names
```

```python
# Load the relevant sheets for analysis
cases_tb = pd.read_excel(file_path, sheet_name='serie casos TB')

# Display the first few rows of each sheet to inspect their structure
cases_tb.head()
```

```python
# Load the relevant sheets for analysis
cases_tb = pd.read_excel(file_path, sheet_name='serie casos TB', skiprows=4, nrows=34)

# Display the first few rows of each sheet to inspect their structure
cases_tb.head()
```

```python
cases_tb.columns = ['State_Short', 'State_Code', 'State_Name'] + [
    f"{year}_{metric}" for year in range(1990, 2024)
    for metric in ['Cases', 'Population', 'Incidence']
]

cases_tb.head()
```

```python
cases_tb_cleaned = cases_tb.dropna(subset=['State_Short', 'State_Code', 'State_Name'])

cases_tb_long = cases_tb.melt(
    id_vars=['State_Short', 'State_Code', 'State_Name'],
    var_name='Year_Metric',
    value_name='Value'
)

cases_tb_long.head()
```

```python
cases_tb_long[['Year', 'Metric']] = cases_tb_long['Year_Metric'].str.split('_', expand=True)
cases_tb_long = cases_tb_long.drop(columns=['Year_Metric'])

cases_tb_long.head()
```

```python
output_csv_path = 'data/serie_historica_tratada_casos.csv'
cases_tb_long.to_csv(output_csv_path, index=False)
```
